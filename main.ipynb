# I. Traitement des données brutes

Lecture des deux csv

```python
import pandas as pd

df_red = pd.read_csv('winequality-red.csv', sep=";")
df_white = pd.read_csv('winequality-white.csv', sep=";")
```

```python
df_red.head()
```

```python
df_white.head()
```

Ajout d'une colonne pour chaque data frame correspondant à la couleur

```python
df_red['color'] = 'red'
df_red.head()
```

```python
df_white['color'] = 'white'
df_white.head()
```

Concaténation des deux data frames en un seul

```python
df = pd.concat([df_red, df_white])
df
```

Transformation de la colonne couleur en dummies

```python
one_hot = pd.get_dummies(df.color, prefix = 'color')
df = pd.concat((df, one_hot), axis=1)
df = df.drop('color', axis=1)

df
```

# II. Statistiques descriptives

Qualité moyenne du vin

```python
df.quality.mean()
```

Diagramme de répartition de la qualité du vin

```python
import matplotlib.pyplot as plt 

plt.title('Répartition de la qualité du vin')
df.quality.hist()
plt.xlabel('Notes')
plt.ylabel('Effectifs')
```

Diagramme de la répartition de la qualité du vin en fonction de si c'est du vin rouge ou du vin blanc

```python
plt.title('Répartition de la qualité du vin')
df_red.quality.hist(alpha = .50)
df_white.quality.hist(alpha = .50)
plt.legend(['vin rouge', 'vin blanc'])
plt.xlabel('Notes')
plt.ylabel('Effectifs')
```

# III. Données d'entraînement

Attribution de X et y
  - Pour y (élément à prédire) : on prend la qualité du vin
  - Pour X : on prend tous les autres critères

```python
X = df.drop('quality', axis = 1)
y = df.quality
```

# IV. Modèle de régression logistique

Entraînement d'un modèle de régression logistique

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
model = LogisticRegression(max_iter=1000000)
model.fit(X_train, y_train)

print('Précision du set d\'entraînement:', model.score(X_train, y_train))
print('Précision du set de test:', model.score(X_test, y_test)) 
```

```python
from sklearn.model_selection import GridSearchCV

gs = GridSearchCV(model, param_grid={}, scoring='accuracy', refit=True, cv=10)
gs.fit(X, y)

gs.cv_results_
```
