# I. Traitement des données brutes

Lecture des deux csv

```python
import pandas as pd

df_red = pd.read_csv('winequality-red.csv', sep=";")
df_white = pd.read_csv('winequality-white.csv', sep=";")
```

```python
df_red.head()
```

```python
df_white.head()
```

Ajout d'une colonne pour chaque data frame correspondant à la couleur

```python
df_red['color'] = 'red'
df_red.head()
```

```python
df_white['color'] = 'white'
df_white.head()
```

Concaténation des deux data frames en un seul

```python
df = pd.concat([df_red, df_white])
df
```

Transformation de la colonne couleur en dummies

```python
one_hot = pd.get_dummies(df.color, prefix = 'color')
df = pd.concat((df, one_hot), axis=1)
df = df.drop('color', axis=1)

df
```

# II. Statistiques descriptives

Qualité moyenne du vin

```python
df.quality.mean()
```

Qualité médiane du vin

```python
df.quality.median()
```

Diagramme de répartition de la qualité du vin

```python
import matplotlib.pyplot as plt 

plt.title('Répartition de la qualité du vin')
df.quality.hist()
plt.xlabel('Notes')
plt.ylabel('Effectifs')
```

Diagramme de la répartition de la qualité du vin en fonction de si c'est du vin rouge ou du vin blanc

# III. Critères de prédiction
Attribution de X et y :
  - Pour y (élément à prédire) : on prend la qualité du vin
  - Pour X : on prend tous les autres critères

```python
X = df.drop('quality', axis = 1)
y = df.quality
```

# IV. Modèle de régression linéaire

Entraînement d'un modèle de régression linéaire

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

model = LinearRegression()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

Entraînement de 500 modèles linéaires

```python
scores = []

for i in range(500):
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
  model = LinearRegression()
  model.fit(X_train, y_train)
  scores.append(model.score(X_test, y_test))
  
plt.title('Répartition des scores du modèle de régression linéaire')
plt.hist(scores)
plt.xlabel('Scores')
plt.ylabel('Effectifs')
```

Prévisions de la qualité du vin sur ce modèle

```python
y_pred = model.predict(X_test)
y_pred
```

Marge d'erreur sur la prévision de la note

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_pred, y_test)
```
